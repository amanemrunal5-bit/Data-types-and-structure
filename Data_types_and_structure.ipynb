Answer 3:-In Python, lists and tuples are both used to store collections of items, but they differ in several important ways. The main difference is that lists are mutable, meaning their elements can be changed, added, or removed after creation, while tuples are immutable, so their contents cannot be modified once defined. Lists use square brackets [ ], whereas tuples use parentheses ( ). Because tuples are immutable, they are generally faster and use less memory compared to lists, making them suitable for fixed data that should not change, such as coordinates or constant values. Lists, on the other hand, offer more built-in methods like append, remove, and sort, making them better for dynamic data that needs frequent updating. Additionally, tuples can be hashed and used as dictionary keys (when containing only immutable items), while lists cannot.

Answer 4 :-In Python, dictionaries store data using a key–value mapping, which means each piece of data is accessed through a unique key rather than a numeric index. Internally, dictionaries use a hash table to organize and retrieve data efficiently. When you create a key, Python computes a hash value for that key using a hashing function. This hash determines where the corresponding value will be stored in memory. Because of this structure, dictionaries provide very fast lookups, usually in constant time, O(1). Keys must be immutable and hashable—such as strings, numbers, or tuples—so that their hash value remains consistent. Values, however, can be of any data type. When you look up a value, Python recalculates the hash of the key, locates its position in the hash table, and returns the associated value. This system allows dictionaries to store data efficiently and retrieve it quickly, even when dealing with large datasets.

Answer 5 :-You might use a set instead of a list in Python when you need to store a collection of unique items and do not care about maintaining order. Sets automatically remove duplicates, making them ideal for tasks like filtering repeated values from data. They also provide much faster membership testing (checking if an item exists), because sets use a hash-based structure that allows lookups in O(1) time, compared to lists which require a slower O(n) search. Additionally, sets are a good choice when you want to perform mathematical operations such as unions, intersections, and differences, all of which are built-in and efficient with sets but not with lists.

Answer 8:-A hash table is a data structure that stores data in key–value pairs and uses a hashing function to determine where each key’s value is stored in memory. The hashing function converts a key (like a string or number) into a hash code—usually a numeric value—that determines the index or “slot” where the data should be placed. This structure allows for very fast data retrieval because the program can jump directly to the location of a value using its hash, rather than searching through the entire dataset.

Answer 10 :-Strings are immutable in Python because once a string is created, its value cannot be changed. This design choice provides several advantages. First, immutability ensures data integrity, preventing accidental modifications that could introduce bugs. Second, it allows strings to be hashable, meaning they can be used as keys in dictionaries or elements in sets, which rely on consistent values for correct hashing. Third, immutable objects are more memory-efficient because Python can safely reuse the same string object in multiple places without creating copies. Finally, immutability simplifies program behavior and optimization, as the interpreter doesn’t need to track changes to string objects. Any operation that seems to modify a string, such as concatenation or replacement, actually creates a new string object, leaving the original unchanged.

Answer 11 :-Dictionaries offer several advantages over lists for certain tasks in Python because of their key–value structure and hash-based implementation. First, dictionaries allow fast lookups, insertions, and deletions, typically in constant time O(1), whereas searching for an item in a list requires linear time O(n). Second, dictionaries let you associate meaningful keys with values, making it easier to organize and access data based on identifiers rather than numeric indexes. Third, dictionaries naturally prevent duplicate keys, ensuring that each key maps to a unique value, which is useful for tasks like counting occurrences or storing unique configurations. Finally, dictionaries support dynamic and flexible data storage, allowing you to quickly update, add, or remove entries without rearranging the rest of the data, unlike lists where modifying structure can be more cumbersome. These features make dictionaries especially advantageous for tasks like fast data retrieval, mapping relationships, or building lookup tables.
